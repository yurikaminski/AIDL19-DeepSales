```python
import numpy as np
import os
from keras.models import Model
from keras import layers
from keras.models import load_model
import pandas as pd
import lib.util as ut
import tensorflow as tf
```

## Load and compile the model to be analyzed

```python
model_path='./trained_models/model_paper/model_25epochs.h5'
model = load_model(model_path)
model.compile(optimizer='Adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

## Some parameters

```python
audio_path = "./data/audiodata"
labels_file = "./data/conflictlevel.csv"

intervals_seconds = 3
sample_rate = 8000
```

## Read the labels and obtain a binary label from target scores

```python
labels_df = pd.read_csv(labels_file, header=None)
labels_df["class"] = labels_df[1].apply(lambda x: 0 if x < 0 else 1)
```

## Read the test split

```python
test_df = pd.read_csv("./data/test_files.csv")
```

## Obtain the samples from test audios. Each audio is divided into 3 second chunks with 50% of overlapping

```python
x_test,y_test=ut.get_chunks_from_raw(audio_path, test_df,intervals_seconds=3,sample_rate=8000 )
num_samples_test = len(x_test)
```

## Build a model to obtain the output of layers of interest

```python
# Layers we want to analyze
num_layer_interest = 15 # maxout_dense_2 layer
layer_outputs = [layer.output for layer in model.layers[num_layer_interest:(num_layer_interest+1)]] 
# Build a model to obtain the output of layers of interest
activation_model = Model(inputs=model.input, outputs=layer_outputs) 
```

## Predict the test samples 

```python
x_arr = np.array(x_test).reshape([num_samples_test,24000,1])
y_arr = np.array(y_test)
activations=activation_model.predict(x_arr)

```

## Create a dataframe with the label and save it to CSV

```python
lastLayer = pd.DataFrame.from_records(activations)
lastLayer['label']=y_arr
lastLayer.to_csv('./trained_models/model_paper/maxout64_layer.csv',sep=';',index=False, encoding='utf-8')
```
